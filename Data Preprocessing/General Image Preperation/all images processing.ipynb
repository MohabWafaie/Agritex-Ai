```python
import cv2
import os
import numpy as np  # Import NumPy library

def canny_filter(image_path, output_path):
    """Applies Canny filter to an image and saves the result.

    Args:
    image_path: Path to the input image.
    output_path: Path to save the output image.
    """
    # Read the image
    image = cv2.imread(image_path)
    # Convert to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    blurred = cv2.GaussianBlur(gray_image, (5, 5), 0)

    # Apply Sobel filter in x and y directions
    canny = cv2.Canny(blurred, 30, 150)

    # Save the filtered image
    cv2.imwrite(output_path, canny.astype(np.uint8))

def process_folder(folder_path, output_folder_path):
  """Processes all images in a folder and its subfolders.

  Args:
    folder_path: Path to the folder to process.
    output_folder_path: Path to save the output images.
  """
  # Create output folder if it doesn't exist
  os.makedirs(output_folder_path, exist_ok=True)

  for filename in os.listdir(folder_path):
    # Get the full path of the file
    file_path = os.path.join(folder_path, filename)
    # Check if it's a regular file
    if os.path.isfile(file_path):
      # Check if it's an image
      if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp', '.tiff', '.ico', '.webp','.tif','JPG')):
        # Generate output path
        output_path = os.path.join(output_folder_path, f"canny_{filename}")
        # Apply Sobel filter
        canny_filter(file_path, output_path)
    # If it's a folder, process it recursively
    elif os.path.isdir(file_path):
      process_folder(file_path, os.path.join(output_folder_path, filename))

# Specify the folder to process and output folder
folder_to_process = r"D:\Projects\Weed Detection\Processed data"
output_folder = r"D:\Projects\Weed Detection\filtered data"

# Process the folder
process_folder(folder_to_process, output_folder)

print("canny filter applied to all images in the folder and its subfolders.")
```

```python
import cv2
import os
import numpy as np  # Import NumPy library

def sobel_filter(image_path, output_path):
    """Applies Sobel filter to an image and saves the result.

    Args:
    image_path: Path to the input image.
    output_path: Path to save the output image.
    """
    # Read the image
    image = cv2.imread(image_path)
    # Convert to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply Sobel filter in x and y directions
    sobel_x = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0)
    sobel_y = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1)

    # Combine the results using the L2 norm
    sobel_combined = cv2.sqrt(sobel_x**2 + sobel_y**2)
    # Normalize the image
    sobel_normalized = sobel_combined / sobel_combined.max() * 255

    # Save the filtered image
    cv2.imwrite(output_path, sobel_normalized.astype(np.uint8))

def process_folder(folder_path, output_folder_path):
  """Processes all images in a folder and its subfolders.

  Args:
    folder_path: Path to the folder to process.
    output_folder_path: Path to save the output images.
  """
  # Create output folder if it doesn't exist
  os.makedirs(output_folder_path, exist_ok=True)

  for filename in os.listdir(folder_path):
    # Get the full path of the file
    file_path = os.path.join(folder_path, filename)
    # Check if it's a regular file
    if os.path.isfile(file_path):
      # Check if it's an image
      if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp', '.tiff', '.ico', '.webp','.tif','JPG')):
        # Generate output path
        output_path = os.path.join(output_folder_path, f"sobel_{filename}")
        # Apply Sobel filter
        sobel_filter(file_path, output_path)
    # If it's a folder, process it recursively
    elif os.path.isdir(file_path):
      process_folder(file_path, os.path.join(output_folder_path, filename))

# Specify the folder to process and output folder
folder_to_process = r"D:\Projects\Weed Detection\Processed data"
output_folder = r"D:\Projects\Weed Detection\filtered data"

# Process the folder
process_folder(folder_to_process, output_folder)

print("Sobel filter applied to all images in the folder and its subfolders.")
```

```python
import cv2
import os
import numpy as np  # Import NumPy library

def high_pass_filter(image_path, output_path, kernel_size=3, blur_kernel_size=3):
  """Applies a high-pass filter to an image with noise reduction and saves the result.

  Args:
    image_path: Path to the input image.
    output_path: Path to save the output image.
    kernel_size: Size of the filter kernel for low-pass filtering (odd number, default 3).
    blur_kernel_size: Size of the Gaussian blur kernel for noise reduction (default 3).
  """
  # Read the image
  image = cv2.imread(image_path)
  # Convert to grayscale
  gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

  # Apply Gaussian blur for noise reduction (optional)
  blurred_image = cv2.GaussianBlur(gray_image, (blur_kernel_size, blur_kernel_size), 0)

  # Create a low-pass filter kernel (average filter)
  kernel = np.ones((kernel_size, kernel_size), np.float64) / (kernel_size * kernel_size)

  # Apply the low-pass filter to the blurred image (or original image if not blurred)
  filtered_low = cv2.filter2D(blurred_image.astype(np.float64), -1, kernel)

  # Subtract the low-pass filtered image from the original image to get high-pass filtered result
  high_passed = blurred_image.astype(np.float64) - filtered_low

  # Apply median filter for noise reduction (optional)
  high_passed = cv2.medianBlur(high_passed.astype(np.uint8), 3)  # Adjust window size as needed

  # Normalize the image (optional)
  # high_passed_normalized = high_passed / high_passed.max() * 255

  # Save the filtered image
  cv2.imwrite(output_path, high_passed)

def process_folder(folder_path, output_folder_path):
  """Processes all images in a folder and its subfolders.

  Args:
    folder_path: Path to the folder to process.
    output_folder_path: Path to save the output images.
  """
  # Create output folder if it doesn't exist
  os.makedirs(output_folder_path, exist_ok=True)

  for filename in os.listdir(folder_path):
    # Get the full path of the file
    file_path = os.path.join(folder_path, filename)
    # Check if it's a regular file
    if os.path.isfile(file_path):
      # Check if it's an image
      if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp', '.tiff', '.ico', '.webp','.tif','JPG')):
        # Generate output path
        output_path = os.path.join(output_folder_path, f"highpass_{filename}")
        # Apply Sobel filter
        high_pass_filter(file_path, output_path)
    # If it's a folder, process it recursively
    elif os.path.isdir(file_path):
      process_folder(file_path, os.path.join(output_folder_path, filename))

# Specify the folder to process and output folder
folder_to_process = r"D:\Projects\Weed Detection\balanced_dataset"
output_folder = r"D:\Projects\Weed Detection\filtered_dataset"

# Process the folder
process_folder(folder_to_process, output_folder)

print("high pass filter applied to all images in the folder and its subfolders.")
```

```python
import cv2
import numpy as np
import os

def detect_green_areas(input_dir, output_dir, min_area_fraction=0.001):
    # Ensure output directory exists
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
        
    counter = 1
    # Loop through every file in the input directory
    for filename in os.listdir(input_dir):

        if filename.endswith(('.png', '.jpg', '.jpeg','.tif','.JPG')):  # You can add more file extensions if needed
            image_path = os.path.join(input_dir, filename)
            
            # Read the image
            image = cv2.imread(image_path)

            # Convert image to HSV
            hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

            # Define the lower and upper bounds for green color
            lower_green = np.array([27, 30, 30])
            upper_green = np.array([90, 255, 255])

            # Create a binary mask where green regions are white
            mask = cv2.inRange(hsv, lower_green, upper_green)

            # Morphological operations to remove noise
            kernel = np.ones((3, 3), np.uint8)
            mask = cv2.erode(mask, kernel, iterations=1)
            mask = cv2.dilate(mask, kernel, iterations=1)
            mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)
            mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
            

            # Calculate minimum area based on image dimensions
            min_area = min_area_fraction * image.shape[0] * image.shape[1]

            # Find connected components and their statistics
            num_labels, labels, stats, centroids = cv2.connectedComponentsWithStats(mask, connectivity=8)

            # Filter out components based on their area
            for i in range(1, num_labels):  # Exclude background label (0)
                if stats[i, cv2.CC_STAT_AREA] < min_area:
                    mask[labels == i] = 0  # Set pixels of small components to 0

            # Create a white background image
            white_background = np.ones_like(image) * 255

            # Merge the original image and the white background using the mask
            result = cv2.bitwise_and(image, image, mask=mask)
            result += cv2.bitwise_and(white_background, white_background, mask=cv2.bitwise_not(mask))

            # Save the result to the output directory
            output_path = os.path.join(output_dir, filename)
            cv2.imwrite(output_path, result)
            print(counter)
            counter += 1

# Paths to the directories
input_directory = r"D:\Projects\Weed Detection\mohab data\to be done\tomato\tomato"
output_directory = r"D:\Projects\Weed Detection\mohab data\to be done\tomato\processed"
# Call the function with min_area_fraction set to 0.001 (adjust as needed)
detect_green_areas(input_directory, output_directory, min_area_fraction=0.009)
```

```python
from PIL import Image
import os

def resize_and_pad_images(input_folder, output_folder, target_size=(224, 224)):
    # Iterate over all files and subdirectories in the input folder
    for root, dirs, files in os.walk(input_folder):
        # Create corresponding subdirectories in the output folder
        relative_path = os.path.relpath(root, input_folder)
        output_subfolder = os.path.join(output_folder, relative_path)
        os.makedirs(output_subfolder, exist_ok=True)

        # Process all files in the current directory
        for filename in files:
            input_filepath = os.path.join(root, filename)
            # Check if the file is an image
            if any(input_filepath.endswith(ext) for ext in ['.png', '.jpg', '.jpeg', '.gif', '.bmp', '.tiff', '.ico', '.webp','.tif','JPG']):
                try:
                    # Open the image file
                    with Image.open(input_filepath) as img:
                        # Calculate aspect ratio
                        aspect_ratio = img.width / img.height
                        # Resize the image while maintaining aspect ratio
                        if aspect_ratio > 1:  # Landscape orientation
                            new_width = target_size[0]
                            new_height = int(new_width / aspect_ratio)
                        else:  # Portrait orientation
                            new_height = target_size[1]
                            new_width = int(new_height * aspect_ratio)
                        img = img.resize((new_width, new_height), Image.LANCZOS)
                        # Create a new blank image with white background
                        new_img = Image.new("RGB", target_size, "white")
                        # Calculate the position to paste the resized image
                        position = ((target_size[0] - new_width) // 2, (target_size[1] - new_height) // 2)
                        # Paste the resized image onto the new image
                        new_img.paste(img, position)
                        # Save the modified image to the output folder
                        output_filepath = os.path.join(output_subfolder, filename)
                        new_img.save(output_filepath, quality=100)
                        print(f"{filename} processed successfully.")
                except Exception as e:
                    print(f"Error processing {filename}: {e}")

# Example usage
input_folder = r"D:\Projects\Weed Detection\Processed data"
output_folder = r"D:\Projects\Weed Detection\resized_dataset"
resize_and_pad_images(input_folder, output_folder)
```
